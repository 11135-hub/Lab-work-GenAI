```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers


def linear_beta_schedule(timesteps):
    beta_start = 1e-4
    beta_end   = 0.02
    return np.linspace(beta_start, beta_end, timesteps)

betas       = linear_beta_schedule(TIMESTEPS)
alphas      = 1.0 - betas
alphas_cumprod_np = np.cumprod(alphas) # Temporary numpy array
alphas_cumprod = tf.constant(alphas_cumprod_np, dtype=tf.float32) # Convert to tf.constant

# Forward process: add noise at timestep t
def q_sample(x0, t, noise=None):
    if noise is None:
        noise = tf.random.normal(shape=tf.shape(x0))

    # Use tf.gather for indexing and reshape for broadcasting
    sqrt_alpha_t_reshaped  = tf.reshape(tf.gather(alphas_cumprod, t) ** 0.5, (-1, 1, 1, 1))
    sqrt_1malpha_t_reshaped = tf.reshape((1.0 - tf.gather(alphas_cumprod, t)) ** 0.5, (-1, 1, 1, 1))

    return sqrt_alpha_t_reshaped * x0 + sqrt_1malpha_t_reshaped * noise

# Custom Layer to wrap tf.tile
class TileLayer(layers.Layer):
    def __init__(self, multiples, **kwargs):
        super().__init__(**kwargs)
        self.multiples = multiples

    def call(self, inputs):
        return tf.tile(inputs, self.multiples)

# ---- Simple Denoising Model (UNet-lite) ----
def build_model(img_size=28, time_emb_dim=32):
    # Image input
    img_input  = keras.Input(shape=(img_size, img_size, 1))
    # Timestep input (as a number 0-999)
    time_input = keras.Input(shape=(1,))

    # Timestep embedding
    t_emb = layers.Dense(time_emb_dim, activation='swish')(time_input)
    t_emb = layers.Dense(time_emb_dim)(t_emb)
    t_emb = layers.Reshape((1, 1, time_emb_dim))(t_emb)
    # Use the custom TileLayer here
    t_emb = TileLayer(multiples=[1, img_size, img_size, 1])(t_emb)  # broadcast

    # Encoder
    x = layers.Conv2D(32, 3, padding='same', activation='swish')(img_input)
    x = layers.concatenate([x, t_emb[..., :32]])

    x = layers.Conv2D(64, 3, padding='same', activation='swish')(x)
    x = layers.Conv2D(64, 3, padding='same', activation='swish')(x)

    # Decoder
    x = layers.Conv2D(32, 3, padding='same', activation='swish')(x)
    output = layers.Conv2D(1, 1, padding='same')(x)  # predict noise

    return keras.Model([img_input, time_input], output)

model = build_model()
model.summary()
```

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers

def linear_beta_schedule(timesteps):
    beta_start = 1e-4
    beta_end   = 0.02
    return np.linspace(beta_start, beta_end, timesteps)

betas       = linear_beta_schedule(TIMESTEPS)
alphas      = 1.0 - betas
alphas_cumprod_np = np.cumprod(alphas) # Temporary numpy array
alphas_cumprod = tf.constant(alphas_cumprod_np, dtype=tf.float32) # Convert to tf.constant
def q_sample(x0, t, noise=None):
    if noise is None:
        noise = tf.random.normal(shape=tf.shape(x0))

    # Use tf.gather for indexing and reshape for broadcasting
    sqrt_alpha_t_reshaped  = tf.reshape(tf.gather(alphas_cumprod, t) ** 0.5, (-1, 1, 1, 1))
    sqrt_1malpha_t_reshaped = tf.reshape((1.0 - tf.gather(alphas_cumprod, t)) ** 0.5, (-1, 1, 1, 1))

    return sqrt_alpha_t_reshaped * x0 + sqrt_1malpha_t_reshaped * noise

# Custom Layer to wrap tf.tile
class TileLayer(layers.Layer):
    def __init__(self, multiples, **kwargs):
        super().__init__(**kwargs)
        self.multiples = multiples

    def call(self, inputs):
        return tf.tile(inputs, self.multiples)

def build_model(img_size=28, time_emb_dim=32):
    # Image input
    img_input  = keras.Input(shape=(img_size, img_size, 1))
    # Timestep input (as a number 0-999)
    time_input = keras.Input(shape=(1,))

    # Timestep embedding
    t_emb = layers.Dense(time_emb_dim, activation='swish')(time_input)
    t_emb = layers.Dense(time_emb_dim)(t_emb)
    t_emb = layers.Reshape((1, 1, time_emb_dim))(t_emb)
    t_emb = TileLayer(multiples=[1, img_size, img_size, 1])(t_emb)  # broadcast

    # Encoder
    x = layers.Conv2D(32, 3, padding='same', activation='swish')(img_input)
    x = layers.concatenate([x, t_emb[..., :32]])

    x = layers.Conv2D(64, 3, padding='same', activation='swish')(x)
    x = layers.Conv2D(64, 3, padding='same', activation='swish')(x)

    # Decoder
    x = layers.Conv2D(32, 3, padding='same', activation='swish')(x)
    output = layers.Conv2D(1, 1, padding='same')(x)  # predict noise

    return keras.Model([img_input, time_input], output)

model = build_model()
model.summary()
```

```python
optimizer = keras.optimizers.Adam(learning_rate=1e-3)

@tf.function
def train_step(x_batch):
    batch_size = tf.shape(x_batch)[0]

    # Random timestep for each image
    t = tf.random.uniform((batch_size,), 0, TIMESTEPS,
                           dtype=tf.int32)

    # Sample random noise
    noise = tf.random.normal(shape=tf.shape(x_batch))

    # Forward process: add noise to images
    # This now assumes q_sample is modified to handle batch inputs
    noisy_imgs = q_sample(x_batch, t, noise)

    # Normalize timestep to 0-1 for model input
    t_norm = tf.cast(t, tf.float32) / TIMESTEPS
    t_input = tf.reshape(t_norm, (-1, 1))

    with tf.GradientTape() as tape:
        predicted_noise = model([noisy_imgs, t_input],
                                 training=True)
        # Loss: predicted noise vs actual noise (MSE)
        loss = tf.reduce_mean((noise - predicted_noise) ** 2)

    grads = tape.gradient(loss, model.trainable_variables)
    optimizer.apply_gradients(
        zip(grads, model.trainable_variables))
    return loss

# Training
dataset = tf.data.Dataset.from_tensor_slices(x_train)\
            .shuffle(1000).batch(BATCH_SIZE)

for epoch in range(EPOCHS):
    epoch_loss = []
    for batch in dataset:
        loss = train_step(batch)
        epoch_loss.append(loss.numpy())
    print(f"Epoch {epoch+1}/{EPOCHS} | Loss: {np.mean(epoch_loss):.4f}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

def p_sample(model, x, t):
    # x shape: (n, 28, 28, 1)
    n = x.shape[0]
    t_norm = np.full((n, 1), t / TIMESTEPS, dtype=np.float32)

    pred_noise = model.predict([x, t_norm], verbose=0)

    alpha_t      = alphas[t]
    alpha_bar_t  = alphas_cumprod[t]
    alpha_bar_t1 = alphas_cumprod[t-1] if t > 0 else 1.0
    beta_t       = betas[t]

    x0_pred = (x - (1 - alpha_bar_t)**0.5 * pred_noise) / alpha_bar_t**0.5
    x0_pred = np.clip(x0_pred, -1, 1)  # stability ke liye

    mean = (alpha_bar_t1**0.5 * beta_t / (1 - alpha_bar_t)) * x0_pred \
         + (alpha_t**0.5 * (1 - alpha_bar_t1) / (1 - alpha_bar_t)) * x

    if t > 0:
        noise = np.random.normal(size=x.shape)
        x = mean + beta_t**0.5 * noise
    else:
        x = mean
    return x

def generate_images(model, n=4):
    x = np.random.normal(size=(n, IMG_SIZE, IMG_SIZE, 1)).astype(np.float32)

    for t in reversed(range(TIMESTEPS)):
        if t % 100 == 0:
            print(f"Denoising step {t}...")
        x = p_sample(model, x, t)

    fig, axes = plt.subplots(1, n, figsize=(n*2, 2))
    for i, ax in enumerate(axes):
        img = (x[i, :, :, 0] + 1) / 2.0
        img = np.clip(img, 0, 1)
        ax.imshow(img, cmap='gray')
        ax.axis('off')
    plt.suptitle("Generated Fashion MNIST Images")
    plt.tight_layout()
    plt.show()

generate_images(model, n=4)
```

```python

# ================================
# 1. Imports
# ================================
import torch
import torch.nn as nn
import torch.optim as optim
from torchvision import datasets, transforms
import matplotlib.pyplot as plt

# ================================
# 2. Config
# ================================
DEVICE = "cuda" if torch.cuda.is_available() else "cpu"
BATCH_SIZE = 64
EPOCHS = 10
LR = 1e-3
TIMESTEPS = 200

print("Using device:", DEVICE)

# ================================
# 3. Dataset (MNIST)
# ================================
transform = transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize((0.5,), (0.5,))  # important for good results
])

dataset = datasets.MNIST(
    root="/kaggle/working",
    train=True,
    download=True,
    transform=transform
)

loader = torch.utils.data.DataLoader(dataset, batch_size=BATCH_SIZE, shuffle=True)

# ================================
# 4. Diffusion Schedule
# ================================
beta_start = 1e-4
beta_end = 0.02

betas = torch.linspace(beta_start, beta_end, TIMESTEPS).to(DEVICE)
alphas = 1. - betas
alpha_hat = torch.cumprod(alphas, dim=0)

# ================================
# 5. Simple Model (CNN)
# ================================
class SimpleCNN(nn.Module):
    def __init__(self):
        super().__init__()

        self.net = nn.Sequential(
            nn.Conv2d(1, 64, 3, padding=1),
            nn.ReLU(),

            nn.Conv2d(64, 64, 3, padding=1),
            nn.ReLU(),

            nn.Conv2d(64, 1, 3, padding=1)
        )

    def forward(self, x, t):
        # (t ignored for simplicity)
        return self.net(x)

model = SimpleCNN().to(DEVICE)
optimizer = optim.Adam(model.parameters(), lr=LR)
loss_fn = nn.MSELoss()

# ================================
# 6. Add Noise (Forward Diffusion)
# ================================
def add_noise(x, t):
    noise = torch.randn_like(x)

    sqrt_alpha_hat = torch.sqrt(alpha_hat[t])[:, None, None, None]
    sqrt_one_minus_alpha_hat = torch.sqrt(1 - alpha_hat[t])[:, None, None, None]

    noisy_x = sqrt_alpha_hat * x + sqrt_one_minus_alpha_hat * noise
    return noisy_x, noise

# ================================
# 7. Training
# ================================
print("Training started...")

for epoch in range(EPOCHS):
    for i, (images, _) in enumerate(loader):

        images = images.to(DEVICE)

        t = torch.randint(0, TIMESTEPS, (images.size(0),)).to(DEVICE)

        noisy_images, noise = add_noise(images, t)

        predicted_noise = model(noisy_images, t)

        loss = loss_fn(predicted_noise, noise)

        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

        if i % 200 == 0:
            print(f"Epoch {epoch+1} | Step {i} | Loss: {loss.item():.4f}")

print("Training finished!")

# ================================
# 8. Sampling with Steps (Blur → Clear)
# ================================
def sample_with_steps(model):
    model.eval()

    x = torch.randn((1, 1, 28, 28)).to(DEVICE)

    # timesteps where we capture output
    save_steps = [199, 150, 100, 50, 20, 0]
    saved_images = []

    with torch.no_grad():
        for t in reversed(range(TIMESTEPS)):

            t_tensor = torch.tensor([t]).to(DEVICE)

            predicted_noise = model(x, t_tensor)

            alpha = alphas[t]
            alpha_hat_t = alpha_hat[t]
            beta = betas[t]

            if t > 0:
                noise = torch.randn_like(x)
            else:
                noise = torch.zeros_like(x)

            # reverse diffusion equation
            x = (1 / torch.sqrt(alpha)) * (
                x - ((1 - alpha) / torch.sqrt(1 - alpha_hat_t)) * predicted_noise
            ) + torch.sqrt(beta) * noise

            if t in save_steps:
                saved_images.append(x.cpu())

    model.train()
    return saved_images

# ================================
# 9. Plot Blur → Clear Progression
# ================================
images = sample_with_steps(model)

titles = [
    "Pure Noise",
    "Very Blurry",
    "Blurry",
    "Shape Forming",
    "Almost Clear",
    "Final Image"
]

plt.figure(figsize=(12, 3))

for i in range(len(images)):
    plt.subplot(1, len(images), i+1)
    plt.imshow(images[i][0][0], cmap="gray")
    plt.title(titles[i])
    plt.axis("off")

plt.show()
```
